```python
import pandas as pd
import requests, lxml, urllib.parse
from bs4 import BeautifulSoup
import re
import time

ads_details=[]
ads_link=[]
ads_price=[]
ads_from=[]
ads_total_review=[]
ads_images=[]
image=[]

# Adding User-agent (default user-agent from requests library is 'python-requests')
# https://github.com/psf/requests/blob/589c4547338b592b1fb77c65663d8aa6fbb7e38b/requests/utils.py#L808-L814
headers = {
    "User-Agent":
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3538.102 Safari/537.36 Edge/18.19582"
}

user_input=input("Enter content to search: ")
# Search query
params = {'q': user_input}

# Getting HTML response
html = requests.get(f'https://www.google.com/search?q=',
                    headers=headers,
                    params=params).text

# Getting HTML code from BeautifulSoup
soup = BeautifulSoup(html, 'lxml')

# Looking for container that has all necessary data findAll() or find_all()DALGre  RnJeZd top pla-unit-title
for container in soup.findAll('div', class_='RnJeZd top pla-unit-title'): 
    #Scraping title
    title = container.text.replace('|',' ')
    ads_details.append(title)
    
    #Scraing price 
    price = soup.find('div', class_='e10twf T4OwTb').text
    ads_price.append(price)
    
    #scrping Ads From
    ad_from = soup.find('span', class_='zPEcBd VZqTOd').text
    ads_from.append(ad_from)
    
    #scrping Ads total review
    total_review = soup.find('span', class_='fl pbAs0b').text
    ads_total_review.append(total_review)
        
    #Creating beginning of the link to join afterwards
    startOfLink = 'https://www.googleadservices.com/pagead'
    #Scraping end of the link to join afterwards
    endOfLink = container.find('a')['href']
    #Combining (joining) relative and absolute URL's (adding begining and end link)
    ad_link = urllib.parse.urljoin(startOfLink, endOfLink)
    ads_link.append(ad_link )
    
    print(f'Ads_title :{title}\nAd_Link : {ad_link}\nAd_Price: {price}\nAds_From:{ad_from}\nTotal_Review :{total_review}\n\n')
    
    
#creating a dataframe using pandas library
Ads_details= pd.DataFrame({'Ads_Details': ads_details,'Ad_link':ads_link,'Ad_Price':ads_price,'Ads_from':ads_from,'Total Review':ads_total_review})

#timestamp so we dont overwrite old CSVs
ts = time.time()

#Saving data in CSV file:
Ads_details.to_csv("Ads_details"+str(ts)+".csv", encoding="utf-8")

```
